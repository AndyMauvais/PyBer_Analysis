```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(x_axis, y_axis, marker = "D", color="green", s=y_axis)
plt.xlabel("Total number of rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title('Average Fare vs Total Rides')
plt.show()
```

```python
y_axis_larger = [size*3 for size in y_axis]

plt.scatter(x_axis, y_axis, s=y_axis_larger)
plt.show()
```

## Object Oriented  Bubble Plotting

```python
#OOP
y_axis_larger = [size * 15 for size in y_axis]
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=y_axis_larger, color= 'xkcd:mint green', alpha=0.8, edgecolor="black", linewidth=2,label="Boston")

ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.invert_yaxis()
ax.set_xlim(0,50)


fig.patch.set_facecolor('xkcd:white')
ax.set_facecolor('xkcd:gray')

fig.set_size_inches(7,5)
fig.set_dpi(100)

plt.show()
```

# Pie Charts

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(
    y_axis,
    labels=x_axis,
    autopct="%.1f%%",
    explode = explode_values,
    colors = colors
)

plt.show()
```

### Object Oriented Pie Plotting

```python

fig, ax = plt.subplots()
ax.pie(
    y_axis,
    labels = x_axis,
    )

fig.set_size_inches(7,5)
fig.set_dpi(100)
plt.show()
```

#### Adding Error bars to a line chart

```python
import statistics
stdev = statistics.stdev(y_axis)
plt.errorbar(x_axis,y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
